```python
import csv
import matplotlib.pyplot as plt
import time
temp = csv.reader(open('대한민국 기온 데이터.csv'))
rain = csv.reader(open('대한민국 강수량 데이터.csv'))
next(temp)
next(rain)

print('--------------------------------------')
print(country + ' 기후 그래프를 살펴봅시다.')
print('--------------------------------------')

time.sleep(2)
print('1. 월별 평균 기온 그래프를 확인하세요.')

x = []
for a in range(12):
  x.append([])
for row in temp :
  if row[2] != '' :
    x[int(row[0].split('-')[1]) - 1].append(float(row[2]))

plt.rc('font', family = 'NanumGothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('월평균 기온')
plt.boxplot(x)
plt.show()

time.sleep(2)
print('2. 월별 강수량 그래프를 확인하세요.')

y = []
for a in range(12):
  y.append([])
for row in rain :
  if row[2] != '' :
    y[int(row[0].split('-')[1]) - 1].append(float(row[2]))
  
plt.rc('font', family = 'NanumGothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('월평균 강수량')
plt.boxplot(y)
plt.show()

time.sleep(2)
print('--------------------------------------')
print(country + ' 은/는 어떤 기후에 해당될까요?.')
print('질문에 Y 또는 N라고 답하세요.')
print('--------------------------------------')

while True : 
  Q1 = input('1. 연 강수량이 500ml 이하로 매우 적은가요? : ')
  if Q1 == 'Y' :
    print(country +'은 건조기후 입니다.')
    print('건조기후는 수분이 부족하여 수목이 자라기 힘듭니다.')
    path = './건조기후.jpg'
    image_pil = Image.open(path)
    image = np.array(image_pil)
    plt.imshow(image)
    plt.show()
    break
  else :
    print('다음 문항에 응답하세요.')
    time.sleep(1)

  Q2 = input('2. 계절의 변화가 뚜렷한가요? : ' )
  if Q2 == 'Y' :
    print('다음 문항에 응답하세요.')
    time.sleep(1)
    Q3 = input('3.가장 추운 달의 평균 기온이 -3도 이하인가요? : ')
    if Q3 == 'Y' :
      print(country +'은 냉대기후 입니다.')
      print('냉대기후는 온대기후와 한대기후 사이에 나타납니다.')
      print('최한월 평균기온이 -3도 이하입니다.')
      path = './냉대기후.jpg'
      image_pil = Image.open(path)
      image = np.array(image_pil)
      plt.imshow(image)
      plt.show()
      break
    else :
      print(country +'은 온대기후 입니다.')
      print('온대기후는 사계절이 뚜렷하며 중위도에 위치합니다.')
      print('다른 기후에 비해 온화합니다.')
      path = './온대기후.jpg'
      image_pil = Image.open(path)
      image = np.array(image_pil)
      plt.imshow(image)
      plt.show()
      break

  time.sleep(1)
  Q4 = input('4. 가장 따뜻한 달의 평균 기온이 10도 이하인가요? : ' )
  if Q4 == 'Y' :
    print(country +'은 한대기후 입니다.')
    print('한대기후는 고위도에 위치하여 저온으로 수목이 생육할 수 없습니다.')
    path = './한대기후.jpg'
    image_pil = Image.open(path)
    image = np.array(image_pil)
    plt.imshow(image)
    plt.show()
  else :
    print(country +'은 열대기후 입니다.')
    print('열대기후는 일년내내 비가 많이 내리고 덥습니다.')
    path = './열대기후.jpg'
    image_pil = Image.open(path)
    image = np.array(image_pil)
    plt.imshow(image)
    plt.show()
```
